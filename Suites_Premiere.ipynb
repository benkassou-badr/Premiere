On considère la suite $U_n$ définie, pour tout entier naturel $n$, par $U_n = n^3 - n + 4$.

### 1) Déterminer les trois premiers termes de la suite

````python
def u(n):
    return n**3 - n + 4

print(f"u(0) = {u(0)}")
print(f"u(1) = {u(1)}")
print(f"u(2) = {u(2)}")
````

### 2) Exprimer $U_{n+1}$ en fonction de $n$



### 3) Montrer que $U_{n+1} - U_n = 3n(n + 1)$



### 4) Étudier le sens de variations de la suite $(U_n)$



### 5) Programme Python

````python
def u(n):
    return n**3 - n + 4

def seuil(A):
    n = 0
    while u(n) < A:
        n = n + 1
    return n

print("Fonction seuil définie avec succès")
````

#### (a) Quel est le rôle de la fonction u placée avant la fonction seuil ?

La fonction `u(n)` calcule le terme général de la suite pour un rang `n` donné.

#### (b) Exécuter la fonction seuil avec différentes valeurs

````python
print(f"seuil(100) = {seuil(100)}")
print(f"u({seuil(100)}) = {u(seuil(100))}")
print()
print(f"seuil(100000) = {seuil(100000)}")
print(f"u({seuil(100000)}) = {u(seuil(100000))}")
print()
print(f"seuil(10**20) = {seuil(10**20)}")
print(f"u({seuil(10**20)}) = {u(seuil(10**20))}")
````

#### (c) Conjecturer la limite éventuelle de la suite $(u_n)$
