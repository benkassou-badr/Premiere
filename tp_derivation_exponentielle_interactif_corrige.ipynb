## 2. Exercice 1 — Dérivation et tangente

**But de l'activité :** écrire des fonctions Python permettant le calcul de taux de variation, de nombres dérivés, du coefficient directeur et de l'ordonnée à l'origine d'une tangente à une courbe.

On considère la fonction $f$ définie sur $\mathbb{R}$ par :

$$
f(x)=e^x-3.
$$


### 2.1 Fonction `f`

Écrire une fonction Python `f(x)` qui renvoie l'image de `x` par la fonction $f$.


```python
from math import exp

def f(x):
    return exp(x) - 3
```

### 2.2 Coefficient directeur

Écrire une fonction `coeff_dir(xA,yA,xB,yB)` qui calcule le coefficient directeur de la droite passant par deux points $A(x_A,y_A)$ et $B(x_B,y_B)$.


```python
def coeff_dir(xA, yA, xB, yB):
    return (yB - yA) / (xB - xA)
```

### 2.3 Taux de variation

Écrire une fonction `taux_variation(f,a,h)` qui renvoie le taux de variation de `f` entre `a` et `a+h`.


```python
def taux_variation(f, a, h):
    return (f(a + h) - f(a)) / h
```

Calculer le taux de variation de $f$ entre $1$ et $1.000001$.

Conjecturer la valeur du nombre dérivé $f'(1)$.


```python
taux_variation(f, 1, 1e-6)
```

### 2.4 Tangente

Écrire une fonction `coeff_tang(f,a)` qui renvoie :
- le coefficient directeur de la tangente à la courbe de `f` au point d’abscisse `a` ;
- l’ordonnée à l’origine de cette tangente.


```python
def coeff_tang(f, a):
    m = taux_variation(f, a, 1e-6)
    b = f(a) - m * a
    return m, b
```

Tester cette fonction pour déterminer l’équation de la tangente à la courbe de $f$ en $0$.


```python
coeff_tang(f, 0)
```

### 2.5 Table de valeurs

La fonction suivante permet d'obtenir une liste de valeurs de la fonction `f`.


```python
def tab_val(f):
    t=[]
    x=0
    for k in range(10):
        t.append(f(x))
        x=x+2
    return t
```

Questions :

Adapter cette fonction pour qu’elle reçoive en argument la valeur initiale `x0`, le pas `p` et le nombre de valeurs `n`.


```python
def tab_val_param(f, x0, p, n):
    t = []
    x = x0
    for k in range(n):
        t.append(f(x))
        x = x + p
    return t
```

## 3. Visualisation graphique de la fonction

Exécuter la cellule suivante pour afficher la courbe de $f(x)=e^x-3$.


```python
import numpy as np
import matplotlib.pyplot as plt

def f_numpy(x):
    return np.exp(x)-3

x = np.linspace(-2, 3, 400)

plt.figure(figsize=(7,4))
plt.plot(x, f_numpy(x), label="f(x)=e^x-3")
plt.axhline(0)
plt.axvline(0)
plt.grid()
plt.legend()
plt.title("Courbe de la fonction exponentielle étudiée")
plt.show()
```

## 4. Tangente interactive

Exécuter les cellules suivantes pour visualiser la tangente à la courbe en un point d’abscisse `a`.

Déplacer le curseur et observer l’effet sur la tangente.


```python
from ipywidgets import interact, FloatSlider

def tangente_interactive(a):
    x = np.linspace(-2, 3, 400)
    y = np.exp(x)-3
    pente = np.exp(a)
    y_a = np.exp(a)-3
    tang = pente*(x-a)+y_a

    plt.figure(figsize=(7,4))
    plt.plot(x, y, label="f(x)=e^x-3")
    plt.plot(x, tang, label="Tangente en a", linestyle="--")
    plt.scatter([a], [y_a])
    plt.axhline(0)
    plt.axvline(0)
    plt.grid()
    plt.legend()
    plt.title("Tangente à la courbe")
    plt.show()

interact(tangente_interactive, a=FloatSlider(value=0, min=-1.5, max=2, step=0.1));
```

## 5. Exercice 2 — Méthode de Newton

**But de l'activité :** approcher la solution d'une équation à l'aide de la méthode de Newton.

On considère encore la fonction

$$
f(x)=e^x-3.
$$

On cherche à résoudre l’équation :

$$
e^x-3=0.
$$


### 5.1 Fonction `etape_Newton`

Écrire une fonction `etape_Newton(f,a)` qui renvoie l’abscisse du point obtenu après une étape de la méthode de Newton.

Rappel :

$$
x_{n+1}=a-\frac{f(a)}{f'(a)}
$$


```python
def etape_Newton(f, a):
    fp = taux_variation(f, a, 1e-6)
    return a - f(a) / fp
```

### 5.2 Itérations successives

La fonction suivante applique plusieurs étapes de Newton.


```python
def appl_Newton(f, a, n):
    t=[a]
    for k in range(n):
        a=etape_Newton(f,a)
        t.append(a)
    return t
```

Tester cette fonction pour `a=1` et `n=10`.


```python
appl_Newton(f, 1, 10)
```

Donner une valeur approchée de la solution $\alpha$ de l’équation $e^x-3=0$.


```python
appl_Newton(f, 1, 10)[-1]
```

## 6. Visualisation interactive de la méthode de Newton

Exécuter les cellules suivantes pour visualiser géométriquement la méthode de Newton.


```python
def newton_graphique(a0, n):
    x = np.linspace(-2, 3, 400)
    y = np.exp(x)-3

    plt.figure(figsize=(8,5))
    plt.plot(x, y, label="f(x)=e^x-3")
    plt.axhline(0)
    plt.axvline(0)

    a = a0
    for k in range(n):
        y_a = np.exp(a)-3
        pente = np.exp(a)
        tang = pente*(x-a)+y_a

        plt.plot(x, tang, linestyle="--")
        plt.scatter([a], [y_a])

        x_suivant = a - y_a/pente
        plt.plot([a, x_suivant], [y_a, 0])
        plt.plot([x_suivant, x_suivant], [0, np.exp(x_suivant)-3], linestyle=":")

        a = x_suivant

    plt.grid()
    plt.legend()
    plt.title("Méthode de Newton")
    plt.show()
```

```python
from ipywidgets import IntSlider

interact(
    newton_graphique,
    a0=FloatSlider(value=1, min=-1, max=2.5, step=0.1),
    n=IntSlider(value=4, min=1, max=8, step=1)
);
```

## 7. Exercice 3 — Approximation de $e$ par Jacques Bernoulli

On considère la suite définie pour tout entier naturel $n\geq1$ par

$$
u_n=\left(1+\frac{1}{n}\right)^n.
$$


### 7.1 Fonction `u`

Écrire une fonction `u(n)` qui calcule le terme de rang `n`.


```python
def u(n):
    return (1 + 1/n) ** n
```

Calculer `u(1000)`.


```python
u(1000)
```

### 7.2 Fonction `precision`

Compléter cette fonction pour qu’elle renvoie le rang `n` ainsi qu’une valeur approchée de `u(n)` lorsque :

$$
|u_n-e|\leq 10^{-p}
$$


```python
from math import e

def precision(p):
    n=1
    while abs(u(n) - e) > 10**(-p):
        n = n + 1
    return n, u(n)
```

Tester par exemple la fonction pour `p=2`, `p=3` ou `p=4`.


```python
precision(2)
precision(3)
```

## 8. Visualisation de la suite de Bernoulli

Exécuter les cellules suivantes pour observer la convergence de la suite vers $e$.


```python
def u_numpy(n):
    return (1+1/n)**n

n_vals = np.arange(1, 101)
u_vals = np.array([u_numpy(n) for n in n_vals])

plt.figure(figsize=(7,4))
plt.plot(n_vals, u_vals, label="u_n")
plt.axhline(np.e, linestyle="--", label="e")
plt.grid()
plt.legend()
plt.title("Convergence de la suite de Bernoulli vers e")
plt.show()
```
