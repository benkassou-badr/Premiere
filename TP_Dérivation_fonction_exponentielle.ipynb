## Exercice 1

**But :** écrire des fonctions Python permettant de calculer taux de variation, nombre dérivé et équation de tangente.

On considère la fonction

$$
f(x)=e^x-3
$$

### 1. Fonction f

Écrire une fonction `f(x)` qui renvoie l'image de `x`.

````python
from math import exp

def f(x):
    return exp(x)-3
````

### 2. Coefficient directeur

Écrire une fonction `coeff_dir` qui calcule le coefficient directeur entre deux points.

````python
def coeff_dir(xA,yA,xB,yB):
    return (yB-yA)/(xB-xA)
````

### 3. Taux de variation

````python
def taux_variation(f,a,h):
    return (f(a+h)-f(a))/h
````

````python
taux_variation(f,1,1e-6)
````

### 4. Tangente

````python
def coeff_tang(f,a,h=1e-6):
    m=taux_variation(f,a,h)
    b=f(a)-m*a
    return m,b
````

````python
coeff_tang(f,0)
````

### 5. Table de valeurs

````python
def tab_val(f):
    t=[]
    x=0
    for k in range(10):
        t.append(f(x))
        x=x+2
    return t
````

## Exercice 2 — Méthode de Newton

On cherche une solution de l'équation

$$
e^x-3=0
$$

````python
def etape_Newton(f,a,h=1e-6):
    fp=taux_variation(f,a,h)
    return a-f(a)/fp
````

````python
def appl_Newton(f,a,n):
    t=[a]
    for k in range(n):
        a=etape_Newton(f,a)
        t.append(a)
    return t

appl_Newton(f,1,10)
````

## Exercice 3 — Approximation de e (Bernoulli)

$$
u_n = (1+\frac{1}{n})^n
$$

````python
def u(n):
    return (1+1/n)**n
````

````python
u(1000)
````

````python
from math import e

def precision(p):
    n=1
    while abs(u(n)-e)>10**(-p):
        n=n+1
    return n,u(n)
````
